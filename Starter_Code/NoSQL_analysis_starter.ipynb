### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query_1 ={'scores.Hygiene':20}

# Use count_documents to display the number of documents in the result
count_1 = establishments.count_documents(query_1)
print(f"There are {count_1} documents with a hygine score of 20")

# Display the first document in the results using pprint
print(f"The first result is:")
pprint(establishments.find_one(query_1))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 ={"$and":[{'LocalAuthorityName':{'$regex':'London'}},{'RatingValue':{'$gte':"4"}}]}

# Use count_documents to display the number of documents in the result
count_2 = establishments.count_documents(query_2)
print(f"There are {count_2} documents with London as the Local Authority and has a RatingValue greater than or equal to 4")

# Display the first document in the results using pprint
print(f"The first result is:")
pprint(establishments.find_one(query_2))
```

```python
# Convert the result to a Pandas DataFrame
df_2 = pd.DataFrame(list(establishments.find(query_2)))
# Display the number of rows in the DataFrame
total_row_2 = len(list(establishments.find(query_2)))
print(f"There are {total_row_2} rows in the DataFrame.")
# Display the first 10 rows of the DataFrame
df_2.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
degree_search = 0.01
latitude = establishments.find_one({'BusinessName':'Penang Flavours'}).get('geocode').get('latitude')
lat_range = [latitude - degree_search, latitude + degree_search]

longitude = establishments.find_one({'BusinessName':'Penang Flavours'}).get('geocode').get('longitude')
long_range = [longitude - degree_search, longitude + degree_search]

# Rating value must equal 5
{"RatingValue":"5"}

# Query
query = {"geocode.latitude": {'$lte':latitude + degree_search}, 
            "geocode.latitude": {'$gte':latitude - degree_search},
                "geocode.longitude": {'$lte':longitude + degree_search},
                    "geocode.longitude": {'$gte':longitude - degree_search},
                      "RatingValue":"5" 
        }

# Sort by hygiene score
sort = [('scores.Hygiene', 1)]
limit = 5

# Print the results
results_3 = establishments.find(query).sort(sort).limit(limit)
pprint(list(results_3))
```

```python
# Convert result to Pandas DataFrame
df_3 = pd.DataFrame(establishments.find(query).sort(sort).limit(limit))
df_3.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match': {'scores.Hygiene':0}}
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': {'$sum': 1 }}}
sort_values = {'$sort': { 'count': -1 }}

pipeline = [match_query, group_query, sort_values]


# Print the number of documents in the result
results_4 = list(establishments.aggregate(pipeline))
print("Number of rows in result: ", len(results_4))
# Print the first 10 results
pprint(results_4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df_4=pd.DataFrame(establishments.aggregate(pipeline))
# Display the number of rows in the DataFrame
total_row_4 = len(list(establishments.aggregate(pipeline)))
# Display the first 10 rows of the DataFrame
print("Top ten local authority areas with a hygiene score of 0 are listed below:")
df_4.head(10)
```
